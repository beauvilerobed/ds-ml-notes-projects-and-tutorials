## 4 The Spectral Representation of a Stationary Process

### 4.1 Complex-Valued Stationary Time Series

The process $\{X_t\}$ is a complex-valued stationary process if $\mathbb{E}|X_t|^2 < \infty$,  
$\mathbb{E}X_t$ is independent of $t$, and $\mathbb{E}(X_{t+h} \overline{X_t})$ is independent of $t$.  
The autocovariance function $\gamma_X(\cdot)$ is

$$
\gamma_X(h) = \mathbb{E}(X_{t+h} \overline{X_t}) - \mathbb{E}(X_{t+h}) \mathbb{E}(\overline{X_t}).
$$

Similarly to the real-valued stationary process, we have the properties of $\gamma_X(h)$:

**Theorem 4.1 (Basic properties of $\gamma_X(\cdot)$):**

1. $\gamma_X(0) \ge 0$  
2. $|\gamma_X(h)| \le \gamma_X(0)$ for all $h$  
3. $\gamma_X(h) = \overline{\gamma_X(-h)}$ for all $h$  
4. $\gamma_X$ is Hermitian and nonnegative definite; that is, a (possibly complex-valued) function $\kappa$ defined on the integers is Hermitian and nonnegative definite if and only if:
   - $\kappa(n) = \overline{\kappa(-n)}$, and  
   - for all positive integers $n$ and complex vectors $a = (a_1, \ldots, a_n)^T \in \mathbb{C}^n$,

$$
\sum_{i=1}^n \sum_{j=1}^n a_i \kappa(i - j) \overline{a_j} \ge 0.
$$


### 4.2 The Spectral Distribution of a Linear Combination of Sinusoids

Consider the following simple complex-valued process:

$$
X_t = \sum_{j=1}^{n} A(\lambda_j) e^{it\lambda_j} = \sum_{j=1}^{n} A(\lambda_j) \{ \cos(\lambda_j t) + i \sin(\lambda_j t) \},
$$

noting that $e^{ix} = \cos(x) + i \sin(x)$, in which $-\pi < \lambda_1 < \lambda_2 < \cdots < \lambda_n = \pi$ and $A(\lambda_1), \ldots, A(\lambda_n)$ are  
uncorrelated complex-valued random coefficients (possibly zero) such that:

$$
\mathbb{E}\{A(\lambda_j)\} = 0, \quad j = 1, \ldots, n,
$$

and

$$
\mathbb{E}\{A(\lambda_j) \overline{A(\lambda_j)}\} = \sigma_j^2, \quad j = 1, \ldots, n.
$$

To check its stationarity, we compute:

$$
\mathbb{E}(X_t) = 0,
$$

and

$$
\mathbb{E}(X_{t+h} \overline{X_t}) = \mathbb{E} \left\{ \sum_{j=1}^{n} A(\lambda_j) e^{i(t+h)\lambda_j} \cdot \sum_{i=1}^{n} \overline{A(\lambda_i)} e^{-it\lambda_i} \right\}
= \sum_{j=1}^{n} \sum_{i=1}^{n} \mathbb{E}\left[ A(\lambda_j) \overline{A(\lambda_i)} \right] e^{i(t+h)\lambda_j} e^{-it\lambda_i}.
$$

Since the $A(\lambda_j)$ are uncorrelated and zero mean:

$$
= \sum_{j=1}^{n} \sigma_j^2 e^{ih\lambda_j}.
$$

Thus, we have a complex-valued stationary process $\{X_t\}$ with autocovariance function:

$$
\gamma_X(h) = \sum_{j=1}^{n} \sigma_j^2 e^{ih\lambda_j} = \int_{(-\pi, \pi]} e^{ih\nu} \, dF(\nu),
$$

where

$$
F(\lambda) = \sum_{j: \lambda_j \le \lambda} \sigma_j^2.
$$

The function $F$ is known as the **spectral distribution function** of $\{X_t\}$.


### Theorem 4.2 (Herglotz)

A complex-valued function $\gamma_X(\cdot)$ defined on the integers is non-negative definite if and only if

$$
\gamma_X(h) = \int_{(-\pi, \pi]} e^{ih\nu} \, dF(\nu), \quad \forall h = 0, \pm1, \pm2, \ldots,
$$

where $F(\cdot)$ is a right-continuous, non-decreasing, bounded function on $[-\pi, \pi]$ and $F(-\pi) = 0$.

- The function $F$ is called the **spectral distribution function** of $\gamma_X$.
- If $F(\lambda) = \int_{-\pi}^{\lambda} f(\nu) \, d\nu$, for $-\pi \le \lambda \le \pi$, then $f$ is called the **spectral density** of $\gamma_X(\cdot)$.

---

### Corollary 4.1

A complex-valued function $\gamma_X(\cdot)$ defined on the integers is the autocovariance function (ACVF) of a stationary process $\{X_t, t = 0, \pm1, \pm2, \ldots\}$ if and only if either:

1.  
$$
\gamma_X(h) = \int_{(-\pi, \pi]} e^{ih\nu} \, dF(\nu), \quad \forall h = 0, \pm1, \pm2, \ldots,
$$  
where $F$ is a right-continuous, non-decreasing, bounded function on $[-\pi, \pi]$ with $F(-\pi) = 0$,  
**or**

2.  
$$
\sum_{i=1}^{n} \sum_{j=1}^{n} a_i \gamma_X(i - j) \overline{a_j} \ge 0
$$  
for all positive integers $n$ and all $a = (a_1, \ldots, a_n)^T \in \mathbb{C}^n$.

---

### Corollary 4.2

An absolutely summable complex-valued function $\gamma(\cdot)$ defined on the integers is the autocovariance function of a stationary process if and only if

$$
f(\lambda) = \frac{1}{2\pi} \sum_{n=-\infty}^{\infty} e^{-in\lambda} \gamma(n) \ge 0, \quad \forall \lambda \in [-\pi, \pi],
$$

in which case $f(\cdot)$ is the **spectral density** of $\gamma(\cdot)$.

Corollary 4.2 provides a way to calculate the spectral density of $\gamma_X(\cdot)$ of a stationary process $\{X_t\}$.

---

### Example 4.1

For white noise $\{W_t\} \sim \text{WN}(0, \sigma^2)$, we have

$$
\gamma_W(h) = \sigma^2 \cdot \mathbb{I}(h = 0).
$$

Its spectral density is then

$$
f_W(\lambda) = \frac{1}{2\pi} \sum_{n=-\infty}^{\infty} e^{-in\lambda} \gamma_W(n) = \frac{1}{2\pi} e^{-i \cdot 0 \cdot \lambda} \sigma^2 = \frac{\sigma^2}{2\pi}.
$$

### 4.3 Spectral Densities and ARMA Processes

**Theorem 4.3**

If $\{Y_t\}$ is any zero-mean, possibly complex-valued stationary process with spectral distribution function $F_Y(\cdot)$, and $\{X_t\}$ is the process

$$
X_t = \sum_{j=-\infty}^{\infty} \psi_j Y_{t-j}, \quad \text{where } \sum_{j=-\infty}^{\infty} |\psi_j| < \infty, \tag{4.1}
$$

then $\{X_t\}$ is stationary with spectral distribution function

$$
F_X(\lambda) = \int_{(-\pi, \lambda]} \left| \sum_{j=-\infty}^{\infty} \psi_j e^{-ij\nu} \right|^2 dF_Y(\nu), \quad -\pi \le \lambda \le \pi.
$$

**Proof.**  
A similar argument to the proof of Proposition 2.2 shows that $\{X_t\}$ is stationary with mean zero and autocovariance function (ACVF)

$$
\gamma_X(h) = \sum_{j=-\infty}^{\infty} \sum_{k=-\infty}^{\infty} \psi_j \overline{\psi_k} \gamma_Y(h - j + k), \quad h = 0, \pm1, \pm2, \ldots
$$

Using the spectral representation of $\gamma_Y(\cdot)$:

$$
\gamma_X(h) = \sum_{j,k} \psi_j \overline{\psi_k} \int_{(-\pi, \pi]} e^{i(h - j + k)\nu} \, dF_Y(\nu)
= \int_{(-\pi, \pi]} \left( \sum_j \psi_j e^{-ij\nu} \right) \left( \sum_k \overline{\psi_k} e^{ik\nu} \right) e^{ih\nu} \, dF_Y(\nu)
$$

$$
= \int_{(-\pi, \pi]} e^{ih\nu} \left| \sum_{j=-\infty}^{\infty} \psi_j e^{-ij\nu} \right|^2 \, dF_Y(\nu),
$$

which completes the proof.

---

**Remark 4.1**

If $\{Y_t\}$ has a spectral density $f_Y(\cdot)$ and $\{X_t\}$ is defined by (4.1), then $\{X_t\}$ also has a spectral density $f_X(\cdot)$ given by:

$$
f_X(\lambda) = \left| \psi(e^{-i\lambda}) \right|^2 f_Y(\lambda),
$$

where

$$
\psi(e^{-i\lambda}) = \sum_{j=-\infty}^{\infty} \psi_j e^{-ij\lambda}.
$$


### Theorem 4.4 (Spectral Density of an ARMA(p, q) Process)

Let $\{X_t\}$ be an ARMA$(p, q)$ process (not necessarily causal or invertible) satisfying

$$
\phi(B)X_t = \theta(B)W_t, \quad \{W_t\} \sim \text{WN}(0, \sigma^2),
$$

where $\phi(z) = 1 - \phi_1 z - \cdots - \phi_p z^p$ and $\theta(z) = 1 + \theta_1 z + \cdots + \theta_q z^q$  
have no common zeros and $\phi(z)$ has no zeros on the unit circle. Then $\{X_t\}$ has spectral density

$$
f_X(\lambda) = \frac{\sigma^2}{2\pi} \cdot \frac{|\theta(e^{-i\lambda})|^2}{|\phi(e^{-i\lambda})|^2}, \quad -\pi \le \lambda \le \pi.
$$

**Proof.**  
The assumption that $\phi(z)$ has no zeros on the unit circle guarantees that $X_t$ can be written as

$$
X_t = \sum_{j=-\infty}^{\infty} \psi_j W_{t-j}, \quad \text{where } \sum_{j=-\infty}^{\infty} |\psi_j| < \infty.
$$

From Example 4.1, $\{W_t\}$ has spectral density $\sigma^2 / (2\pi)$. Then, by Theorem 4.3, $\{X_t\}$ has a spectral density.  

Let $U_t = \phi(B)X_t = \theta(B)W_t$. Applying Theorem 4.3:

$$
f_U(\lambda) = |\phi(e^{-i\lambda})|^2 f_X(\lambda) = |\theta(e^{-i\lambda})|^2 f_W(\lambda).
$$

Since $\phi(e^{-i\lambda}) \ne 0$ for all $\lambda \in [-\pi, \pi]$, we can divide both sides to get:

$$
f_X(\lambda) = \frac{|\theta(e^{-i\lambda})|^2}{|\phi(e^{-i\lambda})|^2} \cdot f_W(\lambda) = \frac{\sigma^2}{2\pi} \cdot \frac{|\theta(e^{-i\lambda})|^2}{|\phi(e^{-i\lambda})|^2}.
$$

### 4.4 Causality, Invertibility, and the Spectral Density of ARMA(p, q)

Consider the ARMA$(p, q)$ process $\{X_t\}$ satisfying

$$
\phi(B)X_t = \theta(B)W_t,
$$

where $\phi(z)\theta(z) \ne 0$ for $|z| = 1$.

Factorizing the polynomials $\phi(\cdot)$ and $\theta(\cdot)$, we can rewrite the defining equations in the form:

$$
\prod_{j=1}^{p} (1 - a_j^{-1} B) X_t = \prod_{j=1}^{q} (1 - b_j^{-1} B) W_t,
$$

where:

- $|a_j| > 1$, for $1 \le j \le r$  
- $|a_j| < 1$, for $r < j \le p$  
- $|b_j| > 1$, for $1 \le j \le s$  
- $|b_j| < 1$, for $s < j \le q$

By Theorem 4.4, the spectral density of $\{X_t\}$ is:

$$
f_X(\lambda) = \frac{\sigma^2}{2\pi} \cdot
\frac{ \prod_{j=1}^{q} |1 - b_j^{-1} e^{-i\lambda}|^2 }
     { \prod_{j=1}^{p} |1 - a_j^{-1} e^{-i\lambda}|^2 }.
$$


Now define

$$
\tilde{\phi}(B) = \prod_{1 \le j \le r} (1 - a_j^{-1} B) \prod_{r < j \le p} (1 - \overline{a}_j B) \tag{4.2}
$$

and

$$
\tilde{\theta}(B) = \prod_{1 \le j \le s} (1 - b_j^{-1} B) \prod_{s < j \le q} (1 - \overline{b}_j B).
$$

Then we have that $\{X_t\}$ is also the ARMA process defined by:

$$
\tilde{\phi}(B) X_t = \tilde{\theta}(B) \widetilde{W}_t,
$$

where

$$
\widetilde{W}_t = 
\frac{
\prod_{r < j \le p} (1 - a_j B)
\prod_{s < j \le q} (1 - b_j^{-1} B)
}{
\prod_{r < j \le p} (1 - a_j^{-1} B)
\prod_{s < j \le q} (1 - b_j B)
} W_t.
$$

Based on Theorem 4.4 again, $\{\widetilde{W}_t\}$ has spectral density

$$
f_{\widetilde{W}}(\lambda) =
\frac{
\left|
\prod_{r < j \le p} (1 - a_j e^{-i\lambda})
\prod_{s < j \le q} (1 - b_j^{-1} e^{-i\lambda})
\right|^2
}{
\left|
\prod_{r < j \le p} (1 - a_j^{-1} e^{-i\lambda})
\prod_{s < j \le q} (1 - b_j e^{-i\lambda})
\right|^2
} \cdot \frac{\sigma^2}{2\pi}.
$$

Since

$$
|1 - \overline{b}_j e^{-i\lambda}| = |1 - b_j e^{i\lambda}| = |e^{i\lambda}| \cdot |b_j - e^{-i\lambda}| = |b_j| \cdot |1 - b_j^{-1} e^{-i\lambda}|,
$$

we can rewrite $f_{\widetilde{W}}(\lambda)$ as:

$$
f_{\widetilde{W}}(\lambda) = 
\left( \prod_{r < j \le p} |a_j|^2 \right)
\left( \prod_{s < j \le q} |b_j|^2 \right)
\cdot \frac{\sigma^2}{2\pi}.
$$

Thus,

$$
\{\widetilde{W}_t\} \sim \text{WN} \left(0, \sigma^2 
\left\{
\prod_{r < j \le p} |a_j|
\right\}^2 
\left\{
\prod_{s < j \le q} |b_j|
\right\}^{-2}
\right).
$$

Noting that both $\tilde{\phi}(z)$ and $\tilde{\theta}(z)$ have no roots in $|z| \le 1$, the process $\{X_t\}$ has the **causal invertible representation**:

$$
\tilde{\phi}(B) X_t = \tilde{\theta}(B) \widetilde{W}_t.
$$

---

### Example 4.3

The ARMA process:

$$
X_t - 2 X_{t-1} = W_t + 4 W_{t-1}, \quad \{W_t\} \sim \text{WN}(0, \sigma^2),
$$

is **neither causal nor invertible**.

Introducing:

- $\tilde{\phi}(z) = 1 - 0.5z$
- $\tilde{\theta}(z) = 1 + 0.25z$

we see that $\{X_t\}$ has the **causal invertible representation**:

$$
X_t - 0.5 X_{t-1} = \widetilde{W}_t + 0.25 \widetilde{W}_{t-1}, \quad \{\widetilde{W}_t\} \sim \text{WN}(0, 4\sigma^2).
$$
