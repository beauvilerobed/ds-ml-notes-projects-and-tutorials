### 3 Autoregressive Moving Average (ARMA) Processes

#### 3.1 Definition

An ARMA($p, q$) process $\{X_t\}$ is a stationary process that satisfies

$$
X_t - \phi_1 X_{t-1} - \cdots - \phi_p X_{t-p} = W_t + \theta_1 W_{t-1} + \cdots + \theta_q W_{t-q}
$$

which also can be written as

$$
\phi(B) X_t = \theta(B) W_t
$$

where

$$
\phi(B) = 1 - \phi_1 B - \cdots - \phi_p B^p, \quad
\theta(B) = 1 + \theta_1 B + \cdots + \theta_q B^q,
$$

and $\{W_t\} \sim WN(0, \sigma^2)$.  
We say $\{X_t\}$ is an ARMA($p, q$) process with mean $\mu_X$ if $\{X_t - \mu_X\}$ is an ARMA($p, q$) process.

**Remark 3.1.** For an ARMA($p, q$) process $\{X_t\}$, we always insist that $\phi_p, \theta_q \ne 0$ and that the polynomials

$$
\phi(z) = 1 - \phi_1 z - \cdots - \phi_p z^p, \quad 
\theta(z) = 1 + \theta_1 z + \cdots + \theta_q z^q
$$

have no common factors. This implies it is not a lower order ARMA model. For example, consider a white noise process $W_t$, we can write $X_t = W_t$ or

$$
(1 - 2B + B^2) X_t = (1 - 2B + B^2) W_t
$$

It is presented as an ARMA(2,2) model, but essentially it is white noise.

**Remark 3.2.** ARMA processes can accurately approximate many stationary processes:

- AR($p$) = ARMA($p$, 0): $\theta(B) = 1$.
- MA($q$) = ARMA(0, $q$): $\phi(B) = 1$.

Further, for any stationary process with autocovariance function (ACVF) $\gamma$, and any $k > 0$, there exists an ARMA process $\{X_t\}$ for which

$$
\gamma_X(h) = \gamma(h), \quad h = 0, 1, \ldots, k
$$

---

#### 3.2 Causality and Invertibility

Recall the definition of causal and invertible. Let $\{X_t\}$ be an ARMA($p, q$) process defined by the equation

$$
\phi(B) X_t = \theta(B) W_t
$$


- $\{X_t\}$ is said to be **causal** (or more specifically, a causal function of $\{W_t\}$) if there exists a sequence of constants $\{\psi_j\}$ such that $\sum_{j=0}^{\infty} |\psi_j| < \infty$ and

  $$
  X_t = \sum_{j=0}^{\infty} \psi_j W_{t-j}, \quad t = 0, \pm1, \ldots \tag{3.1}
  $$

- $\{X_t\}$ is said to be **invertible** (or more specifically, an invertible function of $\{W_t\}$) if there exists a sequence of constants $\{\pi_j\}$ such that $\sum_{j=0}^{\infty} |\pi_j| < \infty$ and

  $$
  W_t = \sum_{j=0}^{\infty} \pi_j X_{t-j}, \quad t = 0, \pm1, \ldots \tag{3.2}
  $$

- Neither causality nor invertibility is a property of $\{X_t\}$ alone, but rather of the relationship between $\{X_t\}$ and $\{W_t\}$.

---

### Theorem 3.1

Let $\{X_t\}$ be an ARMA($p, q$) process. Then $\{X_t\}$ is **causal** if and only if

$$
\phi(z) \ne 0 \quad \text{for all } |z| \leq 1.
$$

The coefficients $\{\psi_j\}$ in (3.1) are determined by the relation

$$
\psi(z) = \sum_{j=0}^{\infty} \psi_j z^j = \frac{\theta(z)}{\phi(z)}, \quad |z| \leq 1.
$$

---

### Proof (Sketch)

First, assume that $\phi(z) \ne 0$ for $|z| \leq 1$. Suppose

$$
\phi(z) = 1 - \phi_1 z - \cdots - \phi_p z^p = \phi_p(z - z_1)(z - z_2) \cdots (z - z_p),
$$

then $|z_i| > 1$ for $i = 1, \ldots, p$. For each $i$,

$$
\frac{1}{z - z_i} = -\frac{1}{1 - z/z_i} = -\sum_{k=1}^{\infty} \left( \frac{z}{z_i} \right)^k, \quad \text{for } |z| < |z_i|.
$$

This implies there exists $\varepsilon > 0$ such that $1/\phi(z)$ has a power series expansion:

$$
\frac{1}{\phi(z)} = \sum_{j=0}^{\infty} \zeta_j z^j \equiv \zeta(z), \quad |z| < 1 + \varepsilon \leq \min_i |z_i|.
$$

Consequently, $\zeta_j (1 + \varepsilon/2)^j \to 0$ as $j \to \infty$, so there exists $K > 0$ such that

$$
|\zeta_j| < K (1 + \varepsilon/2)^{-j}, \quad \forall j = 0, 1, 2, \ldots
$$

In particular,

$$
\sum_{j=0}^{\infty} |\zeta_j| < \infty \quad \text{and} \quad \zeta(z)\phi(z) = 1 \quad \text{for } |z| \leq 1.
$$

Therefore, we can apply the operator...


Apply $\zeta(B)$ to both sides of the equation $\phi(B)X_t = \theta(B)W_t$ to obtain

$$
X_t = \zeta(B)\theta(B)W_t.
$$

Thus, we have the desired representation:

$$
X_t = \sum_{j=0}^{\infty} \psi_j W_{t-j}
$$

where the sequence $\{\psi_j\}$ is determined by $\theta(z)/\phi(z)$.

---

Now, assume that $\{X_t\}$ is causal; i.e., $X_t = \psi(B)W_t$ with $\sum_{j=0}^{\infty} |\psi_j| < \infty$. Then

$$
\theta(B)W_t = \phi(B)X_t = \phi(B)\psi(B)W_t.
$$

If we let $\eta(z) = \phi(z)\psi(z) = \sum_{j=0}^{\infty} \eta_j z^j$, for $|z| \leq 1$, we can rewrite this equation as

$$
\sum_{j=0}^{q} \theta_j W_{t-j} = \sum_{j=0}^{\infty} \eta_j W_{t-j}.
$$

Taking inner products of each side with $W_{t-k}$, we obtain:

- $\eta_k = \theta_k$ for $k = 0, \ldots, q$  
- $\eta_k = 0$ for $k > q$

Hence,

$$
\theta(z) = \eta(z) = \phi(z)\psi(z), \quad |z| \leq 1.
$$

Since $\theta(z)$ and $\phi(z)$ have no common zeros, and since $|\psi(z)| < \infty$ for $|z| \leq 1$, we conclude that $\phi(z)$ **cannot be zero** for $|z| \leq 1$.

---

**Remark 3.3.**  
If $\phi(z) = 0$ for some $|z| = 1$, then there is **no stationary solution** of the ARMA equations:

$$
\phi(B)X_t = \theta(B)W_t.
$$


### Theorem 3.2

Let $\{X_t\}$ be an ARMA($p, q$) process. Then $\{X_t\}$ is **invertible** if and only if

$$
\theta(z) \ne 0 \quad \text{for all } |z| \leq 1.
$$

The coefficients $\{\pi_j\}$ in (3.2) are determined by the relation

$$
\pi(z) = \sum_{j=0}^{\infty} \pi_j z^j = \frac{\phi(z)}{\theta(z)}, \quad |z| \leq 1.
$$

---

### Proof

First, assume that $\theta(z) \ne 0$ for $|z| \leq 1$. By the same argument as in the proof of the previous theorem, $1/\theta(z)$ has a power series expansion:

$$
\frac{1}{\theta(z)} = \sum_{j=0}^{\infty} \eta_j z^j \equiv \eta(z), \quad |z| < 1 + \varepsilon,
$$

for some $\varepsilon > 0$ and $\sum_{j=0}^{\infty} |\eta_j| < \infty$.

Then, applying $\eta(B)$ to both sides of the ARMA equations, we have:

$$
\eta(B)\phi(B)X_t = \eta(B)\theta(B)W_t = W_t.
$$

Thus, we obtain the desired representation:

$$
W_t = \sum_{j=0}^{\infty} \pi_j X_{t-j},
$$

where the sequence $\{\pi_j\}$ is determined by $\phi(z)/\theta(z)$.

---

Conversely, if $\{X_t\}$ is invertible, then

$$
W_t = \sum_{j=0}^{\infty} \pi_j X_{t-j} = \pi(B)X_t
$$

for some $\sum_{j=0}^{\infty} |\pi_j| < \infty$. Then:

$$
\phi(B)W_t = \phi(B)\pi(B)X_t = \pi(B)\phi(B)X_t = \pi(B)\theta(B)W_t.
$$

This leads to:

$$
\sum_{j=0}^{q} \phi_j W_{t-j} = \sum_{j=0}^{\infty} \zeta_j W_{t-j},
$$

where

$$
\zeta(z) = \pi(z)\theta(z) = \sum_{j=0}^{\infty} \zeta_j z^j, \quad |z| \leq 1.
$$

Taking inner products of both sides with $W_{t-k}$, we obtain:

- $\zeta_k = \phi_k$ for $k = 0, \ldots, q$  
- $\zeta_k = 0$ for $k > q$

Hence,

$$
\phi(z) = \zeta(z) = \pi(z)\theta(z), \quad |z| \leq 1.
$$

Since $\theta(z)$ and $\phi(z)$ have no common zeros, and since $|\pi(z)| < \infty$ for $|z| \leq 1$, we conclude that $\theta(z)$ **cannot be zero** for $|z| \leq 1$.

---

**Remark 3.4.**  
If $\{X_t\}$ is a stationary solution of the ARMA equations and if $\phi(z)\theta(z) \ne 0$ for $|z| \leq 1$, then

$$
X_t = \sum_{j=0}^{\infty} \psi_j W_{t-j}.
$$



and

$$
W_t = \sum_{j=0}^{\infty} \pi_j X_{t-j}
$$

where

$$
\sum_{j=0}^{\infty} \psi_j z^j = \frac{\theta(z)}{\phi(z)} \quad \text{and} \quad \sum_{j=0}^{\infty} \pi_j z^j = \frac{\phi(z)}{\theta(z)}, \quad |z| \leq 1.
$$

---

**Remark 3.5.**  
Let $\{X_t\}$ be the ARMA process solving the equations $\phi(B)X_t = \theta(B)W_t$, where  
$\phi(z) \ne 0$ and $\theta(z) \ne 0$ for all $|z| = 1$.

Then there exist polynomials $\tilde{\phi}(z)$ and $\tilde{\theta}(z)$, nonzero for $|z| \leq 1$, of degree $p$ and $q$ respectively, and a new white noise sequence $\{W_t^*\}$ such that $\{X_t\}$ satisfies the **causal and invertible** equations:

$$
\tilde{\phi}(B) X_t = \tilde{\theta}(B) W_t^*.
$$

---

**Remark 3.6. (Uniqueness)**  
If $\phi(z) \ne 0$ for all $|z| = 1$, then the ARMA equations $\phi(B)X_t = \theta(B)W_t$ have the **unique stationary solution**:

$$
X_t = \sum_{j=-\infty}^{\infty} \psi_j W_{t-j}
$$

where $\psi_j$ comes from the expansion of $\theta(z)/\phi(z)$.


### 3.3 Computing the ACVF of an ARMA($p, q$) Process

We now provide three methods for computing the ACVF (autocovariance function) of an ARMA process.  
- The **second** method is the most convenient for obtaining a **closed-form** solution.  
- The **third** method is the most convenient for obtaining **numerical values**.

---

#### 3.3.1 First Method

Since the causal ARMA($p, q$) process $\phi(B)X_t = \theta(B)W_t$ has the representation

$$
X_t = \psi(B)W_t = \sum_{j=0}^{\infty} \psi_j W_{t-j}
$$

where

$$
\psi(z) = \sum_{j=0}^{\infty} \psi_j z^j = \frac{\theta(z)}{\phi(z)}, \quad |z| \leq 1,
$$

the autocovariance function (ACVF) of $\{X_t\}$ is then

$$
\gamma_X(h) = \sigma^2 \sum_{j=0}^{\infty} \psi_j \psi_{j+|h|}.
$$

To determine the coefficients $\psi_j$, we use the method of matching coefficients:

$$
(1 + \psi_1 z + \psi_2 z^2 + \psi_3 z^3 + \psi_4 z^4 + \cdots)(1 - \phi_1 z - \phi_2 z^2 - \cdots - \phi_p z^p)
= (1 + \theta_1 z + \theta_2 z^2 + \cdots + \theta_q z^q)
$$

which yields the following difference equations for $\psi_k$:

\begin{align}
\psi_1 - \phi_1 &= \theta_1 \\
\psi_2 - \phi_2 - \psi_1 \phi_1 &= \theta_2 \\
\psi_3 - \phi_3 - \psi_2 \phi_1 - \psi_1 \phi_2 &= \theta_3 \tag{3.3} \\
\vdots \notag
\end{align}

By defining $\theta_0 = 1$, $\theta_j = 0$ for $j > q$, and $\phi_j = 0$ for $j > p$, we obtain the general recurrence:

For $0 \leq j < \max(p, q + 1)$,

$$
\psi_j - \sum_{0 < k \leq j} \phi_k \psi_{j-k} = \theta_j \tag{3.4}
$$

For $j \geq \max(p, q + 1)$,

$$
\psi_j - \sum_{0 < k \leq p} \phi_k \psi_{j-k} = 0 \tag{3.5}
$$

The general solution of (3.5) can be written in the form:

$$
\psi_n = \sum_{i=1}^{k} \sum_{j=0}^{r_i - 1} \alpha_{ij} n^j \zeta_i^{-n}, \quad n \geq \max(p, q + 1) - p,
$$

where $\zeta_i$ are the reciprocals of the roots of $\phi(z)$ and $r_i$ are their multiplicities.



where $\zeta_i$, $i = 1, \ldots, k$ are the distinct zeros of $\phi(z)$ and $r_i$ is the multiplicity of $\zeta_i$. The $p$ constants $\alpha_{ij}$ and the coefficients $\psi_j$, for $0 \leq j < \max(p, q + 1) - p$, are uniquely determined by the $\max(p, q + 1)$ boundary conditions given in (3.4).

---

### Example 3.1

Consider the ARMA process:

$$
X_t - X_{t-1} + 0.25 X_{t-2} = W_t + W_{t-1}.
$$

We have:

- $\phi(z) = 1 - z + 0.25 z^2$
- $\theta(z) = 1 + z$

The root of $\phi(z)$ is $2$ (since $|2| > 1$) with multiplicity 2, and the root of $\theta(z)$ is $-1$ (since $|-1| = 1$) with multiplicity 1. Therefore, $\{X_t\}$ is **causal** but **not invertible**.

To find the ACVF of $\{X_t\}$, we compute the $\psi_j$ coefficients:

- $\psi_0 = 1$
- $\psi_1 = \phi_1 + \theta_1 = 1 + 1 = 2$
- For $j \geq 2$, the recurrence is:
  
  $$
  \psi_j - \psi_{j-1} + 0.25 \psi_{j-2} = 0
  $$

Rewriting the recurrence:

$$
\psi_j - 0.5 \psi_{j-1} = 0.5(\psi_{j-1} - 0.5 \psi_{j-2})
$$

We observe a geometric structure:

$$
\psi_1 - 0.5 \psi_0 = 1.5
$$

So the recurrence becomes:

$$
\psi_j - 0.5 \psi_{j-1} = 3 \cdot 2^{-j}
$$

This leads to the explicit formula:

$$
\psi_j = (1 + 3j) \cdot 2^{-j}, \quad j = 0, 1, 2, \ldots
$$

Now, using the general solution:

- $r_i - 1 = 2 - 1 = 1$
- $\zeta_i = 2$
- Then,

  $$
  \psi_n = \sum_{j=0}^{1} \alpha_{1j} n^j \cdot 2^{-n}, \quad n \geq \max(p = 2, q + 1 = 2) - p = 0
  $$

Boundary conditions:

- $\psi_0 = 1$
- $\psi_1 = 2$

Solving gives:

- $\alpha_{10} = 1$
- $\alpha_{11} = 3$

Thus:

$$
\psi_j = (1 + 3j) \cdot 2^{-j}, \quad j = 0, 1, 2, \ldots
$$

---

Finally, the ACVF is:

$$
\gamma_X(h) = \sigma^2 \sum_{j=0}^{\infty} \psi_j \psi_{j+|h|}
= \sigma^2 \sum_{j=0}^{\infty} (1 + 3j)(1 + 3j + 3h) \cdot 2^{-2j - h}
= \sigma^2 \cdot 2^{-h} \left( \frac{32}{3} + 8h \right)
$$


### 3.3.2 Second Method

The second method is based on the difference equations for $\gamma_X(k)$, $k = 0, 1, 2, \ldots$, which are obtained by multiplying each side of

$$
\phi(B)X_t = \theta(B)W_t
$$

by $X_{t-k}$ and taking expectations. That is,

- For $0 \leq k < \max(p, q + 1)$:

  $$
  \gamma_X(k) - \phi_1 \gamma_X(k - 1) - \cdots - \phi_p \gamma_X(k - p)
  = \sigma^2 \sum_{k \leq j \leq q} \theta_j \psi_{j-k} \tag{3.6}
  $$

- For $k \geq \max(p, q + 1)$:

  $$
  \gamma_X(k) - \phi_1 \gamma_X(k - 1) - \cdots - \phi_p \gamma_X(k - p) = 0 \tag{3.7}
  $$

The right-hand side of equation (3.6) comes from the representation:

$$
X_t = \sum_{j=0}^{\infty} \psi_j W_{t-j}.
$$

---

The general solution to (3.7) has the form:

$$
\gamma_X(h) = \sum_{i=1}^{k} \sum_{j=0}^{r_i - 1} \beta_{ij} h^j \zeta_i^{-h}, \quad h \geq \max(p, q + 1) - p \tag{3.8}
$$

where:

- $\zeta_i$ are the distinct roots of $\phi(z)$
- $r_i$ are their multiplicities
- The $p$ constants $\beta_{ij}$ and the covariances $\gamma_X(j)$ for $0 \leq j < \max(p, q + 1) - p$ are uniquely determined from the boundary conditions (3.6), after computing $\psi_0, \psi_1, \ldots, \psi_q$ using (3.3).

---

### Example 3.2

Consider Example 3.1 again. From (3.7), we get:

$$
\gamma(k) - \gamma(k - 1) + 0.25 \gamma(k - 2) = 0, \quad k \geq 2
$$

This recurrence has the general solution:

$$
\gamma_X(h) = \sum_{j=0}^{1} \beta_{1j} h^j \cdot 2^{-h}, \quad h \geq 0
$$

---

#### Boundary Conditions from (3.6)

Given $\psi_0 = 1$ and $\psi_1 = 2$:

1. For $k = 0$:

   $$
   \gamma(0) - \gamma(1) + 0.25 \gamma(2) = \sigma^2 (\psi_0 + \psi_1) = 3\sigma^2
   $$

2. For $k = 1$:

   $$
   \gamma(1) - \gamma(0) + 0.25 \gamma(1) = \sigma^2 \psi_0 = \sigma^2
   $$

Now using the general solution:

- $\gamma_0 = \beta_{10}$
- $\gamma_1 = \beta_{10} \cdot 2^{-1} + \beta_{11} \cdot 1 \cdot 2^{-1} = \frac{1}{2}(\beta_{10} + \beta_{11})$
- $\gamma_2 = \beta_{10} \cdot 2^{-2} + \beta_{11} \cdot 2 \cdot 2^{-2} = \frac{1}{4}(\beta_{10} + 2\beta_{11})$

Substitute into the boundary conditions:

1.  
   $$
   \gamma(0) - \gamma(1) + 0.25 \gamma(2) = \beta_{10} - \frac{1}{2}(\beta_{10} + \beta_{11}) + 0.25 \cdot \frac{1}{4}(\beta_{10} + 2\beta_{11}) = 3\sigma^2
   $$

   Simplifying gives:

   $$
   \frac{3}{4} \beta_{10} - \frac{3}{8} \beta_{11} = 3\sigma^2 \Rightarrow 3\beta_{10} - 1.5\beta_{11} = 12\sigma^2
   $$

2.  
   $$
   \gamma(1) - \gamma(0) + 0.25 \gamma(1) = \frac{1}{2}(\beta_{10} + \beta_{11}) - \beta_{10} + 0.25 \cdot \frac{1}{2}(\beta_{10} + \beta_{11}) = \sigma^2
   $$

   Simplifying gives:

   $$
   -\frac{1}{2} \beta_{10} + \frac{5}{8} \beta_{11} = \sigma^2 \Rightarrow -3\beta_{10} + 5\beta_{11} = 8\sigma^2
   $$

Solving the system:

- $3\beta_{10} - 1.5\beta_{11} = 12\sigma^2$
- $-3\beta_{10} + 5\beta_{11} = 8\sigma^2$

Add both equations:

$$
3.5 \beta_{11} = 20\sigma^2 \Rightarrow \beta_{11} = 8\sigma^2
$$

Substitute back:

$$
\beta_{10} = \frac{32\sigma^2}{3}
$$

---

### Final Result

Thus,

$$
\gamma_X(h) = \sigma^2 \cdot 2^{-h} \left( \frac{32}{3} + 8h \right)
$$


### 3.3.3 Third Method

The numerical determination of the autocovariance function $\gamma_X(h)$ from equations (3.6) and (3.7)  
can be carried out readily by first finding $\gamma_X(0), \ldots, \gamma_X(p)$ from the equations with $k = 0, 1, \ldots, p$,  
and then using the subsequent equations to determine $\gamma_X(p + 1), \gamma_X(p + 2), \ldots$ recursively.

**Example 3.3.** Consider Example 3.1. We have:

$$
\gamma(2) - \gamma(1) + 0.25\gamma(2) = 0,
$$

$$
\gamma(0) - \gamma(1) + 0.25\gamma(2) = \sigma^2(\psi_0 + \psi_1),
$$

$$
\gamma(1) - \gamma(0) + 0.25\gamma(1) = \sigma^2\psi_0,
$$

providing $\gamma_X(0) = \frac{32\sigma^2}{3}$, $\gamma_X(1) = \frac{28\sigma^2}{3}$, and $\gamma_X(2) = \frac{20\sigma^2}{3}$.  
Then the higher lag autocovariances can now easily be found recursively from the equations:

$$
\gamma_X(k) = \gamma_X(k - 1) - 0.25\gamma_X(k - 2), \quad k = 3, 4, \ldots
$$

**Example 3.4.** Now, we consider the causal AR(2) process:

$$
(1 - \zeta_1^{-1} B)(1 - \zeta_2^{-1} B)X_t = W_t, \quad |\zeta_1|, |\zeta_2| > 1, \ \zeta_1 \ne \zeta_2.
$$

Then,

$$
\phi_1 = \zeta_1^{-1} + \zeta_2^{-1},
$$

$$
\phi_2 = -\zeta_1^{-1} \zeta_2^{-1}.
$$

Based on (3.8), we have:

$$
\gamma_X(h) = \sum_{i=1}^{2} \beta_{i1} \zeta_i^{-h}, \quad h \ge 0.
$$

Boundary conditions provide:

$$
\gamma_X(0) - \phi_1 \gamma_X(1) - \phi_2 \gamma_X(2) = \sigma^2,
$$

$$
\gamma_X(1) - \phi_1 \gamma_X(0) - \phi_2 \gamma_X(1) = 0.
$$

Tedious calculation yields that:

$$
\gamma_X(h) = \frac{\sigma^2 \zeta_1^2 \zeta_2^2}{(\zeta_1 \zeta_2 - 1)(\zeta_2 - \zeta_1)} \left( \frac{\zeta_1^{1 - h}}{\zeta_1^2 - 1} - \frac{\zeta_2^{1 - h}}{\zeta_2^2 - 1} \right), \quad h \ge 0.
$$


```r
# Define the autocorrelation function
rho <- function(h, z1, z2) {
  rho0 <- z1 / (z1^2 - 1) - z2 / (z2^2 - 1)
  res <- (z1^(1 - h) / (z1^2 - 1) - z2^(1 - h) / (z2^2 - 1)) / rho0
  return(res)
}

# Set up plotting parameters
par(mfrow = c(3, 1))        # 3 plots in one column
par(mar = c(4, 4, 2, 0.5))  # Margins
par(bg = "white")           # Background color

# Sequence of lags
h <- seq(0, 20, 1)

# Plot 1: z1 = 2, z2 = 5
plot(h, rho(h, 2, 5), type = "o", col = "blue",
     xlab = "Lag", ylab = expression(rho[X](h)),
     ylim = c(-1, 1))
segments(-1, 0, 21, 0, lty = 2)

# Plot 2: z1 = -10/9, z2 = 2
plot(h, rho(h, -10/9, 2), type = "o", col = "blue",
     xlab = "Lag", ylab = expression(rho[X](h)),
     ylim = c(-1, 1))
segments(-1, 0, 21, 0, lty = 2)

# Plot 3: z1 = 10/9, z2 = 2
plot(h, rho(h, 10/9, 2), type = "o", col = "blue",
     xlab = "Lag", ylab = expression(rho[X](h)),
     ylim = c(-1, 1))
segments(-1, 0, 21, 0, lty = 2)
```
